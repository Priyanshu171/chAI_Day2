```python
for feature in numerical_features:
    if df[feature].isnull().sum()>0:
        print(f"No. of missing values in {feature}: {df[feature].isnull().sum()}")
```

First, we will deal with the missing values of numerical features 
we have various methods for this, some of them are:
replace by mean, median, mode
replace by ffill(forward fill) i.e replace by previous value (useful in continuous distribution)
replace by bfill (backward fill) i.e replace by next value

```python
df['LotFrontage'].describe()
```

```python
df['LotFrontage'].replace(np.nan,(df.LotFrontage.median()+np.random.randint(-10,10)),inplace=True)
# we can add a bit of noise while replacing nan with median for a better distribution
```

```python
df['MasVnrArea'].replace(np.nan,df.MasVnrArea.mean(),inplace=True)
```

```python
df['GarageYrBlt'].fillna(df.GarageYrBlt.median(),inplace=True)
```

```python
for feature in categorical_features:
    if df[feature].isnull().sum():
        print(f"No. of missing values in {feature}: {df[feature].isnull().sum()}")
```

```python
df['Alley'].value_counts()
```

As the number of Nan in alley is too large, we will replace them by Missing and then perform one hot encoding where missing will
be mapped to 0

```python
for feature in categorical_features:
    print(f"""value counts for {feature}: 
{df[feature].value_counts()}
             """)
```

Now, we will convert categorical features into numeric according to the required technique
if the categories are of equal importance like gender, create two separate columns and fill the ones accordingly, this is called as one hot encoding. 
if the categories are of different importance like low, medium and high, we can number them as 1,2,3 respectively 

Generally, we convert all the text to lower for more simplification but in this case, it won't be required

```python
one_hot_cols = ['MSZoning','Street','Alley','RoofStyle','MasVnrType','Foundation','Electrical']
for feature in one_hot_cols:
    if df[feature].isnull().sum()>0:
        print(f"No. of missing values in {feature}: {df[feature].isnull().sum()}")
```

```python
for feature in one_hot_cols:
    if df[feature].isnull().sum()>0:
        df[feature].fillna('missing',inplace=True)
```

```python
dummies = pd.get_dummies(df[one_hot_cols])
print(dummies.shape)
dummies.columns
```

```python
missing_cols = ['Alley_missing','MasVnrType_missing','Electrical_missing']
dummies.drop(missing_cols,axis=1,inplace=True)
```

```python
merged = pd.concat([df,dummies],axis=1)
```

```python
merged.drop(one_hot_cols,axis=1,inplace=True)
merged
```

In this way, we encoded the categorical features using dummies and the missing values will be simply mapped to 0, because we dropped the missing columns

Same can be done using sklearn preprocessing OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
le = LabelEncoder()
```

```python
dummies2 = ohe.fit_transform(df[one_hot_cols])
```

```python
dummies2.shape
```

Now, we will deal with the categorical features which has categories of different importance 

```python
one_hot_cols
```

```python
imp_based_features = ['LotShape','LandContour','Utilities','LandSlope','ExterQual','ExterCond','BsmtQual','BsmtCond',
                      'BsmtExposure','HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond','PoolQC']
for feature in imp_based_features:
    if merged[feature].isnull().sum()>0:
        print(f"No. of missing values in {feature}: {merged[feature].isnull().sum()}")
```

```python
for feature in imp_based_features:
    merged[feature].fillna("missing",inplace=True)
```

```python
for feature in imp_based_features:
    print(f"""value counts for {feature}:
{merged[feature].value_counts()}
    """)
```

```python
lotshape_dict = {'IR3':1,'IR2':2,'IR1':3,'Reg':4}
landcontour_dict = {'Low':1,'HLS':2,'Bnk':3,'Lvl':4}
utilities_dict = {'NoSewa':1,'NoSeWa':1,'AllPub':2}
LandSlope_dict = {'Sev':1,'Mod':2,'Gtl':3}
qual_dict = {'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5,'missing':0,'No':0,'Mn':1,'Av':2}

merged['LotShape'].replace(lotshape_dict,inplace=True)
merged['LandContour'].replace(landcontour_dict,inplace=True)
merged['Utilities'].replace(utilities_dict,inplace=True)
merged['LandSlope'].replace(LandSlope_dict,inplace=True)
# df[4:].replace(qual_dict,inplace=True)

for feature in imp_based_features[4:]:
    merged[feature].replace(qual_dict,inplace=True)
```

```python
len(final_cols)
```

```python
final_df = merged[final_cols]
```

```python
final_df.head()
```

```python
for col in final_df.columns:
    print(f"missing values in {col}: {final_df[col].isnull().sum()}")
```

```python
for col in final_df.columns:
    print(f"""value counts in {col}:
{final_df[col].value_counts()}
""")
```

```python
for col in final_df.columns:
    print(f"""value counts in {col}:
{final_df[col].value_counts()}
""")
```

Feature Engineering:
we will now convert numerical features into some standard form by scaling them
there are various methods to achive the same like:
1. divide by x.max value
2. x-x.min/x.max-x.min (normalization)
3. x-x.mean/x.std (standardization)

else, we can also map values using a particular function like:
1. np.log (logarithm)
2. 1/x (reciprocal)
3. np.exp(x) (exponentiate)

after mapping values using the above function, we plot them and check whether we observe a particular type of distribution is obtained or not such as normal distribution (bell curve)

```python
num_cols = ['LotFrontage','LotArea','MasVnrArea','BsmtFinSF1','BsmtFinSF2'] 
```

```python
final_df.loc[:,'LotFrontage'] = (final_df['LotFrontage']-final_df['LotFrontage'].min())/(final_df['LotFrontage'].max()-final_df['LotFrontage'].min())

"""
we can use MinMaxScaler from sklearn.preprocessing
form its object and then 
df[col].fit_transfrom()
"""
```

```python
final_df.loc[:,'LotArea'] = (final_df['LotArea']- final_df['LotArea'].mean())/final_df['LotArea'].std()

"""
we can use StandardScaler from sklearn.preprocessing to achieve the same
"""
```
